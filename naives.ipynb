```python
def import_data():
    #import data either by gentext or pandas
    data=np.genfromtxt('../input/data2_19.csv',dtype=str)
    data=data[1:]

    test=np.genfromtxt('../input/test2_19.csv',dtype=str)
    test=test[1:]
    
    return data, test 
```

```python
#flattening array into 2dimensions
def get_arr(data):
    
    arr=[]
    j=0
    while(j < len(data)):
        i = data[j]
        temp=[]
        temp.append(int(i[1]))
        temp.append(int(i[3]))
        temp.append(int(i[5]))
        temp.append(int(i[7]))
        temp.append(int(i[9]))
        temp.append(int(i[11]))
        temp.append(int(i[13]))
        arr.append(temp)
        j+=1
        
    return np.asarray(arr)
```

```python
arr = get_arr(data)
arr2 = get_arr(test)
```

```python
a=arr[:,0]
unique, counts = np.unique(a, return_counts=True)
unique,counts
```

```python
def get_apriori(counts, unique):
    #aprior distributon, ,very usfull in probabilistic models
    apriori=[]
    total=np.sum(counts)
    j=0
    while(j < len(counts)) :
        i = counts[j]
        apriori.append((i+1)/(total+len(unique)))
        j+=1
    return apriori
```

```python
apriori = get_apriori(counts, unique)
```

```python
def get_xy(arr, arr2):
    X_train=arr[:,1:]
    y_train=arr[:,0]

    X_test=arr2[:,1:]
    y_test=arr2[:,0]
    return X_train, X_test, y_train, y_test
```

```python
X_train, X_test, y_train, y_test = get_xy(arr, arr2)
```

```python
def prob(x,data,clas):
    #probability of a class
    temp=data[arr[:,0]==clas]
    prob=1
    i=0
    while(i < (temp.shape[1])):
        prob=prob*(((list(temp[:,i]).count(x[i]))+1)/(len(list(temp[:,i]))+5))
        #temp=temp[temp[:,i]==x[i]]
        i+=1
    return prob 
```

```python
def get_pred(X_test, X_train):
    y_pred=[]
    j=0
    while(j  < len(X_test)):
        i=X_test[j]
        p0=prob(i,X_train,0)*apriori[0]
        p1=prob(i,X_train,1)*apriori[1]

        if(p0>p1): y_pred.append(0)
        else: y_pred.append(1)
        j+=1
    return y_pred
```

```python
y_pred = get_pred(X_test, X_train)
```
